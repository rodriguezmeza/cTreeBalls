```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
n=9 # número de multipolos

expData = []
expData2 = []
expData3 = []

#y1="/home/eladio/Documentos/nside_8094/Output_tree-omp-sincos_nres13r000_zs9" #dirección del output
y1="../../Output" #dirección del output
y2="../../Output" #dirección del output
y3="../../Eladio/Output_nside1024_tree-omp-sincos" #dirección del output
# para construir la parte real de la se
for j in range(1,n+1):
    expData3.append(np.loadtxt(y3+"/"+"histZetaM_cos_"+ str(j) +".txt")+np.loadtxt(y3+"/"+"histZetaM_sin_"+ str(j) +".txt"))

for j in range(1,n+1):
    expData2.append(np.loadtxt(y2+"/"+"histZetaM_2_cos_"+ str(j) +".txt")+np.loadtxt(y2+"/"+"histZetaM_2_sin_"+ str(j) +".txt"))    
    
for j in range(1,n+1):
    expData.append(np.loadtxt(y1+"/"+"histZetaM_cos_"+ str(j) +".txt")+np.loadtxt(y1+"/"+"histZetaM_sin_"+ str(j) +".txt"))    
#    expData.append(np.loadtxt(y1+"/"+"histZetaM_1_cos_"+ str(j) +".txt")+np.loadtxt(y1+"/"+"histZetaM_1_sin_"+ str(j) +".txt"))    
    
#    expData.append(np.loadtxt(y1+"/"+"histZetaM_cos_"+ str(j) +".txt")+np.loadtxt(y1+"/"+"histZetaM_sin_"+ str(j) +".txt"))
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(tight_layout=True,figsize=(20, 15))
gs = gridspec.GridSpec(3, 3)
#ax.set_xscale("log")
formatter=mpl.ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
#plt.yscale("log")

# Para m=0
ax = fig.add_subplot(gs[0, :])
ax.plot(range(len(expData[0].flatten())),expData[0].flatten(),color="blue",linestyle="-",alpha=1,label="m=0")
ax.set_ylabel(r"$ \zeta_0$",fontsize=35)
ax.set_xlim(0,400)
#ax.set_ylim(1e-10,5e-7)
ax.set_ylim(1e-10,5e-7)
ax.tick_params(labelsize = 25)
ax.legend(fontsize=25)
ax.set_yscale("log")
ax.axes.xaxis.set_ticklabels([])
# Para m=1
ax = fig.add_subplot(gs[1, :])
ax.plot(range(len(expData[1].flatten())),expData[1].flatten(),color="red",linestyle="-",alpha=1,label="m=1")
ax.set_ylabel(r"$ \zeta_1$",fontsize=35)
ax.set_xlim(0,400)
ax.set_ylim(1e-11,5e-7)
ax.tick_params(labelsize = 25)
ax.legend(fontsize=25)
ax.set_yscale("log")
ax.axes.xaxis.set_ticklabels([])
# Para m=2
ax = fig.add_subplot(gs[2, :])
ax.plot(range(len(expData[2].flatten())),expData[2].flatten(),color="green",linestyle="-",alpha=1,label="m=2")
ax.set_xlabel(r"triangle index",fontsize=30)
ax.set_ylabel(r"$ \zeta_2$",fontsize=35)
#ax.set_title("Full sky, Nside=4096, z=2.0548, 108 realizations, m=0",fontsize=30)
ax.set_xlim(0,400)
#ax.set_ylim(1e-10,5e-7)
ax.set_ylim(0.2e-10,5e-7)
ax.tick_params(labelsize = 25)
ax.legend(fontsize=25)
ax.set_yscale("log")

   
fig.align_labels() 
#plt.savefig("flatten.png",dpi=300)
plt.show()
```
